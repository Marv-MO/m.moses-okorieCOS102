```python
from tabulate import tabulate

Name = ["Evelyn", "Jessica", "Somto", "Edith", "Lisa", "Madonna", "Waje", "Tola", "Aisha", "Latifa"]
Age = [17, 16, 17, 18, 16, 18, 17, 20, 19, 17]
Height = [5.5, 6, 5.4, 5.9, 5.6, 5.5, 6.1, 6.0, 5.7, 5.5]
Scores = [80, 85, 70, 60, 76, 66, 87, 95, 50, 49]

table = zip(Name, Age, Height, Scores)
print(tabulate(table, headers=["Name", "Age", "Height", "Scores"], tablefmt="grid"))
```
