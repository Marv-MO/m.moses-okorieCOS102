```python
# Simple Interest
P = float(input('Enter Principal: '))
R = float(input('Enter Rate: '))
T = float(input('Enter Time: '))
print('Principal = ', P)
print('Rate = ', R)
print('Time = ', T)
A = float(P) * (1 + (float(R) / 100) * float(T))
print('Simple Interest = ', A)

```

```python
# Compound Interest
P = float(input('Enter Principal: '))
R = float(input('Enter Rate: '))
T = float(input('Enter Time: '))
n = float(input('Enter number of times interest applied per time period: '))
print('Principal = ', P)
print('Rate = ', R)
print('Time = ', T)
print('n = ', n)
A = float(P) * (1 + (float(R) / 100) / float(n)) ** (float(n) * float(T))
print('Compound Interest = ', A)

```

```python
# Annuity plan
PMT = float(input('Enter Payment: '))
R = float(input('Enter Rate: '))
T = float(input('Enter Time: '))
n = float(input('Enter n: '))
print('Payment = ', PMT)
print('Rate = ', R)
print('Time = ', T)
print('n = ', n)
A = float(PMT) * ((1 + (float(R)) / float(n)) ** (float(n) * float(T)) - 1) / ((float(R)) / float(n))
print('Annuity plan = ', A)
```
