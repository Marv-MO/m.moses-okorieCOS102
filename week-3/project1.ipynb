```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# Load the dataset
try:
    df = pd.read_csv('GIG-logistics.csv')  # Assuming columns are 'Name' and 'Department'
except FileNotFoundError:
    print("Error: GIG-logistics.csv file not found.")
    exit()

def verify_employee():
    name = name_entry.get().strip()
    dept = dept_entry.get().strip()

    # Check if employee exists
    employee = df[(df['Name'].str.lower() == name.lower()) & 
                  (df['Department'].str.lower() == dept.lower())]
    
    if not employee.empty:
        # Get other members in the department
        other_members = df[(df['Department'].str.lower() == dept.lower()) & 
                           (df['Name'].str.lower() != name.lower())]
        
        if not other_members.empty:
            members_str = ", ".join(other_members['Name'].tolist())
        else:
            members_str = "None"
        
        messagebox.showinfo("Success", 
                            f"Welcome, {name}!\nOther members in {dept} department: {members_str}")
    else:
        messagebox.showerror("Error", "Employee not found. Please check your name and department.")

# GUI setup
root = tk.Tk()
root.title("Employee Verification")

tk.Label(root, text="Name:").pack(pady=5)
name_entry = tk.Entry(root)
name_entry.pack(pady=5)

tk.Label(root, text="Department:").pack(pady=5)
dept_entry = tk.Entry(root)
dept_entry.pack(pady=5)

tk.Button(root, text="Verify", command=verify_employee).pack(pady=15)

root.mainloop()
```
