```python
import tkinter as tk
from tkinter import messagebox

class DeliveryService:
    def __init__(self, root):
        self.root = root
        self.root.title("Delivery Charge Calculator")
        self.root.geometry("400x300")

        # --- UI Elements ---
        tk.Label(root, text="Select Location:").pack(pady=5)
        self.location_var = tk.StringVar(value="PAU")
        tk.Radiobutton(root, text="PAU", variable=self.location_var, value="PAU").pack()
        tk.Radiobutton(root, text="Epe", variable=self.location_var, value="Epe").pack()

        tk.Label(root, text="Enter Package Weight (kg):").pack(pady=10)
        self.weight_entry = tk.Entry(root)
        self.weight_entry.pack()

        self.calc_button = tk.Button(root, text="Calculate Price", command=self.calculate_price)
        self.calc_button.pack(pady=20)

    def calculate_price(self):
        try:
            location = self.location_var.get()
            weight = float(self.weight_entry.get())
            price = 0

            if location == "PAU":
                if weight >= 10:
                    price = 2000
                else:
                    price = 1500
            elif location == "Epe":
                if weight >= 10:
                    price = 5000
                else:
                    price = 4000

            messagebox.showinfo("Delivery Quote", f"The delivery charge to {location} is: N{price}")
            
        except ValueError:
            messagebox.showerror("Error", "Please enter a valid numerical weight.")

# --- Main Execution ---
if __name__ == "__main__":
    root = tk.Tk()
    app = DeliveryService(root)
    root.mainloop()
```
